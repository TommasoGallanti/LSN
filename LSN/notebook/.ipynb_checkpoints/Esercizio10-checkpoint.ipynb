### NON INTERACTING 10.1

```python
def plot_loss(filename, title):
    generation, mean_loss, best_loss = np.loadtxt(filename, usecols=(0,1,2), unpack=True)
    plt.figure(figsize=(10,8))
    plt.plot(generation, mean_loss, color="blue", label="first half mean loss")
    plt.scatter(generation, best_loss, s=7, marker="*", color="red", label="best loss")
    plt.xlabel("Generations")
    plt.ylabel("Loss")
    plt.xscale('log')  # scala logaritmica
    plt.yscale('log')  # scala logaritmica
    plt.grid()  
    plt.legend()
    plt.title(title)
    plt.show()
    print("best loss = ", best_loss[-1])
```

```python
def plot_trip(filename,trip):
    x,y = np.loadtxt(filename, usecols=(0,1), unpack = True)
    plt.figure(figsize=(10,8))
    plt.plot(x,y,label="last generation best trip",color="blue")
    plt.scatter(x,y,label="cities",marker="*",color="red",lw=2)
    plt.scatter(x[0],y[0],color="green",lw=3,label="start")
    plt.scatter(x[-1],y[-1],color="green",lw=3)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid()
    plt.title(trip)
    plt.show();
```

```python
for i in range(0,7):
    plot_loss("../c++/es 10/10.1/SOURCE/Mean_loss_"+str(i+1)+".dat","GA loss function "+str(i+1))
```

```python
for i in range(0,7):
    plot_trip("../c++/es 10/10.1/SOURCE/final_trip_"+str(i+1)+".dat","FINAL TRIP "+str(i+1))
```
