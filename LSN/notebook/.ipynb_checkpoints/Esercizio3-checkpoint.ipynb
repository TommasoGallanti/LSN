## <span style="color:blue">  Numerical exercises 3</span>

### Exercise 03.1 <span style="color:red"> Plain vanilla option pricing </span>



By using the following parameters: 
- asset price at $t=0$: $S(0)=100$  prezzo iniziale
- delivery time: $T=1$
- strike price: $K=100$  prezzo pattuito
- risk-free interest rate: $r=0.1$   tasso interesse
- volatility: $\sigma=0.25$

compute at time $t=0$ **via Monte Carlo** the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$:

<span style="color:red">Make four pictures for the estimation of the European call-option prices, $C[S(0),0]$ (direct and discretized), and put-option prices, $P[S(0),0]$ (direct and discretized), with their uncertainties </span> with a large number of asset prices at time $t=T$, say $M\ge 10^4$, as a function of the number of blocks, $N$</font>. <span style="color:blue"> As usual, in computing statistical uncertainties, use data blocking.</span>

### For comparison, Black-Scholes analytic solution:

```python
call_scholes = 14.975790778311286
put_scholes = 5.4595325819072364
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
plt.rcParams.update({'font.size': 15})
```

```python
data = np.loadtxt("../c++/es 3/Call_direct.dat")
means_call_direct = data[:, 0]  
stds_call_direct = data[:, 1]  
```

```python
data = np.loadtxt("../c++/es 3/Call_discr.dat")
means_call_discr = data[:, 0]  
stds_call_discr = data[:, 1]  
```

```python
data = np.loadtxt("../c++/es 3/Put_direct.dat")
means_put_direct = data[:, 0]  
stds_put_direct = data[:, 1]  
```

```python
data = np.loadtxt("../c++/es 3/Put_discr.dat")
means_put_discr = data[:, 0]  
stds_put_discr = data[:, 1]  
```

- I will use 100 blocks with 100 steps each

- Sampling **directly** the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$ in uno step

```python
x = np.linspace(0,100,100)
fig = plt.figure(figsize = (10,6) )
plt.errorbar(x,means_call_direct,yerr=stds_call_direct, label="simulation",marker="*", color = "grey")
plt.hlines(call_scholes,0,100,color="red", label = "theoretical value",linewidth=3)
plt.legend()
plt.xlabel('#blocks')
plt.ylabel('Call_option')
plt.grid(True)
plt.title("Simulation vs theory (direct method)")

plt.show()

print("<call_direct> = ", round(means_call_direct[-1],3), "+/-", round(stds_call_direct[-1],3))
print("confidence level = ", round((means_call_direct[-1] - call_scholes)/stds_call_direct[-1],3), " sigma" )
```

```python
x = np.linspace(0,100,100)
fig = plt.figure(figsize = (10,6) )
plt.errorbar(x,means_put_direct,yerr=stds_put_direct, label="simulation",marker="*", color = "grey")
plt.hlines(put_scholes,0,100,color="red", label = "theoretical value",linewidth=3)
plt.legend()
plt.xlabel('#blocks')
plt.ylabel('Put_option')
plt.grid(True)
plt.title("Simulation vs theory (direct method)")

plt.show()

print("<put_direct> = ", round(means_put_direct[-1],3), "+/-", round(stds_put_direct[-1],3))
print("confidence level = ", round((means_put_direct[-1] - put_scholes)/stds_put_direct[-1],3), " sigma" )
```

- Sampling the discretized $GBM(r,\sigma^2)$ path of the asset price
dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$

```python
x = np.linspace(0,100,100)
fig = plt.figure(figsize = (10,6) )
plt.errorbar(x,means_call_discr,yerr=stds_call_discr, label="simulation",marker="*", color = "purple")
plt.hlines(call_scholes,0,100,color="red", label = "theoretical value",linewidth=3)
plt.legend()
plt.xlabel('#blocks')
plt.ylabel('Call_option')
plt.grid(True)
plt.title("Simulation vs theory (discret method)")

plt.show()

print("<call_discret> = ", round(means_call_discr[-1],3), "+/-", round(stds_call_discr[-1],3))
print("confidence level = ", round((means_call_discr[-1] - call_scholes)/stds_call_discr[-1],3), " sigma" )
```

```python
x = np.linspace(0,100,100)
fig = plt.figure(figsize = (10,6) )
plt.errorbar(x,means_put_discr,yerr=stds_put_discr, label="simulation",marker="*", color = "purple")
plt.hlines(put_scholes,0,100,color="red", label = "theoretical value",linewidth=3)
plt.legend()
plt.xlabel('#blocks')
plt.ylabel('Put_option')
plt.grid(True)
plt.title("Simulation vs theory (discret method)")

plt.show()

print("<put_discret> = ", round(means_put_discr[-1],3), "+/-", round(stds_put_discr[-1],3))
print("confidence level = ", round((means_put_discr[-1] - put_scholes)/stds_put_discr[-1],3), " sigma" )
```

**comparison between the methods:**

```python
x = np.linspace(0,100,100)
fig = plt.figure(figsize = (10,6) )
plt.errorbar(x,means_put_discr,yerr=stds_put_discr, label="simulation_discrete",marker="*", color = "purple")
plt.errorbar(x,means_put_direct,yerr=stds_put_direct, label="simulation_direct",marker="*", color = "grey")
plt.hlines(put_scholes,0,100,color="red", label = "theoretical value",linewidth=3)
plt.legend(loc="lower right")
plt.xlabel('#blocks')
plt.ylabel('Put_option')
plt.grid(True)
plt.title("Simulation vs theory")

plt.show()
```

```python
x = np.linspace(0,100,100)
fig = plt.figure(figsize = (10,6) )
plt.errorbar(x,means_call_discr,yerr=stds_call_discr, label="simulation_discrete",marker="*", color = "purple")
plt.errorbar(x,means_call_direct,yerr=stds_call_direct, label="simulation_direct",marker="*", color = "grey")
plt.hlines(call_scholes,0,100,color="red", label = "theoretical value",linewidth=3)
plt.legend()
plt.xlabel('#blocks')
plt.ylabel('Call_option')
plt.grid(True)
plt.title("Simulation vs theory")

plt.show()
```
